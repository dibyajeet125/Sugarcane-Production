```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv("List of Countries by Sugarcane Production.csv")
```

```python
df
```

```python
df.info()
```

```python
# Dropping the Unnamed: 0 column
df = df.drop("Unnamed: 0", axis = 1)
```

```python
# Changing the data type of production from object to int
df.isnull().sum()
```

```python
# Dropping the rows with null values
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
# Replcaing "."/"," with ""/"."  respectively
df["Production (Tons)"] = df["Production (Tons)"].str.replace(".","")
df["Production per Person (Kg)"] = df["Production per Person (Kg)"].str.replace(".","").str.replace(",",".")
df["Acreage (Hectare)"] = df["Acreage (Hectare)"].str.replace(".","")
df["Yield (Kg / Hectare)"]= df["Yield (Kg / Hectare)"].str.replace(".","").str.replace(",",".")
```

```python
df
```

```python
# Renaming the columns
df.rename(columns= {"Production (Tons)": "Production(Tons)"}, inplace = True)
df.rename(columns= {"Production per Person (Kg)": "Production_per_person(Kg)"}, inplace = True)
df.rename(columns= {"Acreage (Hectare)": "Acreage(Hectare)"}, inplace = True)
df.rename(columns= {"Yield (Kg / Hectare)": "Yield(Kg/Hectare)"}, inplace = True)
```

```python
# Changing the data type of production from object to float
df["Production(Tons)"] = df["Production(Tons)"].astype(float)
df["Production_per_person(Kg)"] = df["Production_per_person(Kg)"].astype(float)
df["Acreage(Hectare)"] = df["Acreage(Hectare)"].astype(float)
df["Yield(Kg/Hectare)"] = df["Yield(Kg/Hectare)"].astype(float)
```

```python
df.info()
```

```python
# Descrptive statistics of numerical columns
df.describe().T
```

```python
# Descriptive statistics of categorical columns
df.describe(include = "object").T
```

#### Univariate analysis

```python
# Counting the number of countries in each continent
df["Continent"].value_counts()
```

```python
df["Continent"].value_counts().plot(kind = "bar")
```

```python
# Plotting the distribution of numerical columns
plt.figure(figsize = (10,10))
plt.subplot(2,2,1)
sns.distplot(df["Production(Tons)"])
plt.title("Production(Tons)")
plt.subplot(2,2,2)
sns.distplot(df["Production_per_person(Kg)"])
plt.title("Production_per_person(Kg)")
plt.subplot(2,2,3)
sns.distplot(df["Acreage(Hectare)"])
plt.title("Acreage(Hectare)")
plt.subplot(2,2,4)
sns.distplot(df["Yield(Kg/Hectare)"])
plt.title("Yield(Kg/Hectare)")
plt.show()
```

#### Bivariate analysis

```python
# Sugarcane production in each country
df_bi = df[["Country","Production(Tons)"]].set_index("Country")
```

```python
df_bi
```

```python
plt.figure(figsize = (8,8))
df_bi["Production(Tons)"].plot(kind="pie", autopct="%.2f%%", labels=df_bi.index)
plt.title("Sugarcane production in each country")
```

```python
import matplotlib.ticker as mticker

df[["Country","Production(Tons)"]].set_index("Country").sort_values("Production(Tons)", ascending = False).head(15).plot(kind = "bar")
plt.ylabel("Production(Tons)")

# Format the y-axis to show regular numbers instead of scientific notation
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
```

```python
# Land acreage in top 15 country
df[["Country","Acreage(Hectare)"]].set_index("Country").sort_values("Acreage(Hectare)", ascending = False).head(15).plot(kind = "bar")
plt.ylabel("Acreage(Hectare)")

# Format the y-axis to show regular numbers instead of scientific notation
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
```

```python
# Top 15 country yield(Kg/Hectare)
df[["Country","Yield(Kg/Hectare)"]].set_index("Country").sort_values("Yield(Kg/Hectare)", ascending = False).head(15).plot(kind = "bar")
plt.ylabel("Yield(Kg/Hectare)")

# Format the y-axis to show regular numbers instead of scientific notation
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
```

```python
# Sugarcane production/person in top 15 country
df[["Country","Production_per_person(Kg)"]].set_index("Country").sort_values("Production_per_person(Kg)", ascending = False).head(15).plot(kind = "bar")
plt.ylabel("Production_per_person(Kg)")

# Format the y-axis to show regular numbers instead of scientific notation
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
```
